```python
df_las_selected = df_las[["RDEP", "RSHA", "RMED", "GR", "NPHI", "RHOB"]]
```

```python
df_las_selected.isna().sum()
```

```python
df_las_data_dropped = df_las_selected.dropna(subset=["RDEP", "RSHA", "RMED", "GR", "NPHI", "RHOB"], axis=0, how="any")
```

```python
df_las_data_dropped
```

```python
df_las_data_dropped.describe()
```

```python
#filter data 
df_filter = df_las_data_dropped.loc[(df_las_data_dropped.GR > 0) & (df_las_data_dropped.GR <= 250)]
df_filter1 = df_filter.loc[(df_filter.RHOB >= 1) & (df_filter.RHOB <= 3)]

```

```python
#calculate the shale volume
#make a copy of the filtered data
df = df_filter1.copy()
df["VSH"] = (df.GR - df.GR.min()) / (df.GR.max() - df.GR.min()) 
```

```python
df.head(2)
```

```python
#rename index
df_idx = df.rename_axis("DEPTH").reset_index()
```

```python
df_idx.head(2)
```

```python
#plotting with the help of matplotlib i will be writing a function for it
# "RDEP", "RSHA", "RMED", "GR", "NPHI", "RHOB"
def plot_logs(logs):
    logs.sort_values(by="DEPTH")
    top = logs.DEPTH.min()
    bot = logs.DEPTH.max()
    f, ax = plt.subplots(nrows = 1, ncols = 6, figsize=(12,8))
    ax[0].plot(logs.RDEP, logs.DEPTH, color="red")
    ax[1].plot(logs.RSHA, logs.DEPTH, color="green")
    ax[2].plot(logs.RMED, logs.DEPTH, color="yellow")
    ax[3].plot(logs.GR, logs.DEPTH, color="blue")
    ax[4].plot(logs.NPHI, logs.DEPTH, color="orange")
    ax[5].plot(logs.RHOB, logs.DEPTH, color="c")
    for i in range(len(ax)):
        ax[i].set_ylim(top, bot)
        ax[i].invert_yaxis()
        ax[i].grid()
    ax[0].set_xlabel("RDEP")
    ax[0].set_xlim(logs.RDEP.min(), logs.RDEP.max())
    ax[0].set_ylabel("Depth (ft)")
    ax[1].set_xlabel("RSHA")
    ax[1].set_xlim(logs.RSHA.min(), logs.RSHA.max())
    ax[2].set_xlabel("RMED")
    ax[2].set_xlim(logs.RMED.min(), logs.RMED.max())
    ax[3].set_xlabel("GR")
    ax[3].set_xlim(logs.GR.min(), logs.GR.max())
    ax[4].set_xlabel("NPHI")
    ax[4].set_xlim(logs.NPHI.min(), logs.NPHI.max())
    ax[5].set_xlabel("RHOB")
    ax[5].set_xlim(logs.RHOB.min(), logs.RHOB.max())

    ax[1].set_ythicklabels([]);
    ax[2].set_ythicklabels([]);
    ax[3].set_ythicklabels([]);
    ax[4].set_ythicklabels([]);
    ax[5].set_ythicklabels([]);

    f.subtitle("Well log analysis using python by Bello Hadi", fontsize=14, y=0.94)
```

```python
plot_logs(df_idx)
```
